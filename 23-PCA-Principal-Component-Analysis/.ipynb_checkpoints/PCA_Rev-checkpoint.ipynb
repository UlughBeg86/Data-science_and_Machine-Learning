```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/cancer_tumor_data_features.csv")
```

```python
df.head()
```

```python
sns.heatmap(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
scaled_X.shape
```

```python
scaled_X.mean()
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
covariance_matrix
```

```python
covariance_matrix.shape
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values.shape
```

```python
eigen_vectors.shape
```

```python
# PCA N feature space ---> N PC space
# How many PC to choose ? N = 2
```

```python
num_components = 2
```

```python
 np.argsort([2, 1, 3])
```

```python
sorted_key
```

```python
np.argsort(eigen_values)
```

```python
sorted_key =  np.argsort(eigen_values)[::-1][:num_components]
```

```python
sorted_key
```

```python
eigen_vectors[:, sorted_key].shape
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:, sorted_key]
```

```python
eigen_values
```

```python
eigen_vectors.shape
```

```python
# original data --> project --> eigen_vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
principal_components.shape
```

```python
scaled_X.shape
```

```python
eigen_vectors.shape
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.scatter(principal_components[:, 0], principal_components[:, 1])
```

```python
df = pd.read_csv("../DATA/cancer_tumor_data_features.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.shape
```

```python
df.shape
```

```python
scaled_X[:4].shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_model.fit(scaled_X)
```

```python
pca_model.transform(scaled_X)
```

```python
pca_model.transform(scaled_X).shape
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8, 5), dpi=200)
plt.scatter(pc_results[:, 0], pc_results[:, 1])
```

```python
cancer_dictionary["target"]
```

```python
plt.figure(figsize=(5, 4), dpi=100)
plt.scatter(pc_results[:, 0], pc_results[:, 1], c=cancer_dictionary["target"])
```

```python
pca_model.components_
```

```python
pca_model.components_.shape
```

```python
sns.heatmap(pca_model.components_)
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=["PC1", "PC2"], columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20, 4), dpi=200)
sns.heatmap(df_comp, annot=True)
```

```python
pca_model.explained_variance_ratio_
```

```python
pca_model.explained_variance_ratio_.sum()
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30
```

```python
pca_30
```

```python
pca_30.get_covariance
```

```python
explained_variance = []

for n in range(1, 30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(15,7), dpi=100)
plt.plot(range(1, 30), explained_variance)
plt.xlabel("Num of Components")
plt.ylabel("Variance Explained")
```
