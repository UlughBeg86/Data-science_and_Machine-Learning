```python
dbscan = DBSCAN(eps=2)
display_categories(dbscan, two_blobs_outliers)
```

```python
np.sum(dbscan.labels_ == -1)
```

```python
np.sum(dbscan.labels_ == -1)*100/len(dbscan.labels_)
```

```python
np.unique(dbscan.labels_)
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001, 10, 1000):
    
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    
    perc_outlier = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    
    outlier_percent.append(perc_outlier)

```

```python
sns.lineplot(x=np.linspace(0.001, 10, 1000), y=number_of_outliers)
plt.xlim(0, 2)
```

```python
dbscan = DBSCAN(eps=7)
display_categories(dbscan, two_blobs_outliers)
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001, 7, 1000):
    
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    
    perc_outlier = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    
    outlier_percent.append(perc_outlier)

```

```python
sns.lineplot(x=np.linspace(0.001, 7, 1000), y=outlier_percent)
plt.xlim(0, 2)
plt.ylim(0, 10)
plt.ylabel("percent of points classified as Outlier")
```

```python
dbscan = DBSCAN(eps=1)
display_categories(dbscan, two_blobs_outliers)
```

```python
outlier_percent = []
number_of_outliers = []

for eps in np.linspace(0.001, 1, 1000):
    
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    
    perc_outlier = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    
    outlier_percent.append(perc_outlier)

```

```python
dbscan = DBSCAN(eps=1)
display_categories(dbscan, two_blobs_outliers)
```

```python
outlier_percent = []
number_of_outliers = []

for n in np.arange(1, 100):
    
    dbscan = DBSCAN(min_samples=n)
    dbscan.fit(two_blobs_outliers)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    
    perc_outlier = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    
    outlier_percent.append(perc_outlier)

```

```python
sns.lineplot(x=np.arange(1, 100), y=outlier_percent)
plt.xlabel("Min Number of samples")
plt.ylabel("Pct of points classified as an outlier")
#plt.xlim(0, 2)
#plt.ylim(0, 10)
plt.ylabel("percent of points classified as Outlier")
```

```python
dbscan = DBSCAN(eps=0.75, min_samples=1)
display_categories(dbscan, two_blobs_outliers)
```
