## Part2. Clusters and Dendrograms

```python
len(df)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
#scale_df
cluster_labels = model.fit_predict(scale_df)
```

```python
cluster_labels
```

```python
cluster_labels.shape
```

```python
np.unique(cluster_labels)
```

```python
plt.figure(figsize=(16, 6), dpi=200)
sns.scatterplot(data=df, x="mpg", y="weight", hue=cluster_labels, palette="viridis")
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
#scale_df
cluster_labels = model.fit_predict(scale_df)
```

```python
plt.figure(figsize=(16, 6), dpi=200)
sns.scatterplot(data=df, x="mpg", y="weight", hue=cluster_labels, palette="viridis")
```

```python
# What is the max distance possible?
len(scale_df.columns)
```

```python
scale_df.columns
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scale_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
linkage_matrix.shape
```

```python
cluster_labels
```

```python
plt.figure(figsize=(20, 10))
dendro = dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20, 10))
dendro = dendrogram(linkage_matrix, truncate_mode="lastp", p=10)
```

```python
plt.figure(figsize=(20, 10))
dendro = dendrogram(linkage_matrix, truncate_mode="level", p=3)
```
