### Part2: Model

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
# We have a single stump
model = AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test, predictions))
```

```python
error_rates = []

for i in range(1, 96):
    model = AdaBoostClassifier(n_estimators=i)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    err = 1 - accuracy_score(y_test, preds)
    
    error_rates.append(err)
```

```python
plt.plot(range(1, 96), error_rates)
```

```python
model
```

```python
model.feature_importances_
```

```python
model.feature_importances_.shape
```

```python
feats = pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=["Importance"])
```

```python
feats
```

```python
feats.transpose()
```

```python
import_feats = feats[feats["Importance"] > 0]
```

```python
import_feats
```

```python
plt.figure(figsize=(16, 7), dpi=200)
sns.barplot(data=import_feats, x=import_feats.index, y="Importance")
plt.xticks(rotation=90)
```

### Gradient Boosting

```python
df = pd.read_csv("C://Users//rd//Documents//UDEMY/DATA/mushrooms.csv")
df.shape
```

```python
df.head()
```

```python
X = df.drop("class", axis=1)
```

```python
X.columns.shape
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators":[50, 100],
             "learning_rate": [0.1, 0.05, 0.2],
             "max_depth":[3, 4, 5]}
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid = GridSearchCV(estimator=gb_model, param_grid=param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
predictions = grid.predict(X_test)
```

```python
predictions
```

```python
grid.best_estimator_
```

```python
# We got the default values
grid.best_params_
```

```python
print(classification_report(y_test, predictions))
```

```python
grid.best_estimator_.feature_importances_
```

```python
grid.best_estimator_.feature_importances_.shape
```

```python
feat_import = grid.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index=X.columns,
                       data=feat_import,
                       columns=["Importance"])
```

```python
imp_feat
```

```python
type(imp_feat)
```

```python
imp_feat = imp_feat[imp_feat["Importance"] > 0.0005]
```

```python
imp_feat
```

```python
imp_feat.shape
```

```python
sns.barplot(data=imp_feat.sort_values("Importance"), x=imp_feat.index, y="Importance")
```

```python
plt.figure(figsize=(16, 7), dpi=200)
sns.barplot(data=imp_feat.sort_values("Importance"), x=imp_feat.index, y="Importance")
plt.xticks(rotation=90)
```
