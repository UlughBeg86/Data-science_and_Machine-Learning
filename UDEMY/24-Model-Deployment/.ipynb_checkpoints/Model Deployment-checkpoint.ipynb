```python
import pandas as pd
```

```python
df = pd.read_csv("../DATA/Advertising.csv")
```

```python
df.head()
```

```python
df
```

```python
df.shape
```

```python
df.describe()
```

```python
X = df.drop("sales", axis=1) 
y = df["sales"]
```

```python
from sklearn.model_selection import train_test_split
#TRAIN / Validation / Hold out test
# 70%  /  15%  / 15% ==100

# 1st split - 70/30
# 2nd split 50/50
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
X_train.shape
```

```python
X_validation, X_holdout_test, y_validation, y_holdout_test = train_test_split(X_test, y_test, test_size=0.50, random_state=101)
```

```python
len(X)
```

```python
len(X_validation)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
len(X)
```

# Model Training

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=3, random_state=101)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
validation_predictions = model.predict(X_validation)
```

```python
validation_predictions
```

```python
validation_predictions.shape
```

```python
mean_absolute_error(y_validation, validation_predictions)
```

```python
mean_squared_error(y_validation, validation_predictions)**0.5
```

```python
df.describe()["sales"]
```

```python
model = RandomForestRegressor(n_estimators=30, random_state=101)
```

```python
model.fit(X_train, y_train)
```

```python
validation_predictions = model.predict(X_validation)
```

```python
validation_predictions
```

```python
mean_absolute_error(y_validation, validation_predictions)
```

```python
mean_squared_error(y_validation, validation_predictions)**0.5
```

## Final performance metrics (HOLDOUT SET)

```python
holdout_preditions = model.predict(X_holdout_test)
```

```python
mean_absolute_error(y_holdout_test, holdout_preditions)
```

```python
mean_squared_error(y_holdout_test, holdout_preditions)**0.5
```

```python
final_model = RandomForestRegressor(n_estimators=30, random_state=101)
```

```python
final_model.fit(X, y)
```

```python
import joblib
```
